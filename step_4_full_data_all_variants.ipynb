```python
INPUT_FILE = "all_model_variants.csv"
OUTPUT_FILE = "full_data_all_variants.csv"

variants_df = pd.read_csv(INPUT_FILE)
model_url_list = variants_df["Model URL"].to_list()
len(model_url_list)
```

```python
from bs4 import BeautifulSoup
import requests
import os

def get_model_data(site_url):
    
    response = requests.get(site_url)
    site_response = response.text
    soup = BeautifulSoup(site_response, "html.parser")

    # find Title
    if soup.find('h1', class_="newstitle").text.strip() != None:
        car_title = soup.find('h1', class_="newstitle").text.strip()

    model_box = soup.find('div', class_='modelbox')
    result_dict = {}

    # Model box
    if not model_box:
        print('Error: could not find model box')
    else:
        # initialize dictionary to store results
        #result_dict = {}

        # find Body style
        body_style_elem = model_box.find('b', text='Body style:')
        if body_style_elem:
            result_dict['Body style'] = body_style_elem.next_sibling.strip()

        # find Segment
        segment_elem = model_box.find('b', text='Segment:')
        if segment_elem:
            result_dict['Segment'] = segment_elem.next_sibling.strip()

        # find Infotainment
        infotainment_elem = model_box.find('b', text='Infotainment:')
        if infotainment_elem:
            result_dict['Infotainment'] = infotainment_elem.next_sibling.strip()

        # find Production years
        prod_years_elem = model_box.find('b', text='Production years:')
        if prod_years_elem:
            result_dict['Production years'] = prod_years_elem.next_sibling.strip()

        result_dict['Title'] = car_title

    # Description box
    description = soup.find('div', class_='modelcontainer')
    if not description:
        print('Error: could not find description box')
    else:
        description_text = description.text.strip()
        result_dict['Description'] = description_text

    # Engine specifics
    engine_data = soup.find('div', class_='enginedata')
    if not engine_data:
        print('Error: could not find engine specifics box')
    else:
        # find Engine Specs title
        engine_specs_elem = engine_data.find('th', class_='title')
        if engine_specs_elem:
            result_dict['Engine specs title'] = engine_specs_elem.text.split('SPECS - ')[1]

        # find Cylinders
        cylinders_elem = engine_data.find('td', text='Cylinders:')
        if cylinders_elem:
            result_dict['Cylinders'] = cylinders_elem.find_next_sibling('td').text.strip()

        # find Displecement
        displacement_elem = engine_data.find('td', text='Displacement:')
        if displacement_elem:
            result_dict['Displacement'] = displacement_elem.find_next_sibling('td').text.strip()

        # find Power
        power_elem = engine_data.find('td', text='Power:')
        if power_elem:
            result_dict['Power'] = power_elem.find_next_sibling('td').text.strip()

        # find Torque
        torque_elem = engine_data.find('td', text='Torque:')
        if torque_elem:
            result_dict['Torque'] = torque_elem.find_next_sibling('td').text.strip()

        # find Fuel System
        fuel_system_elem = engine_data.find('td', text='Fuel System:')
        if fuel_system_elem:
            result_dict['Fuel System'] = fuel_system_elem.find_next_sibling('td').text.strip()

        # find Fuel
        fuel_elem = engine_data.find('td', text='Fuel:')
        if fuel_elem:
            result_dict['Fuel'] = fuel_elem.find_next_sibling('td').text.strip()

        # find Fuel capacity
        fuel_capacity_elem = engine_data.find('td', text='Fuel capacity:')
        if fuel_capacity_elem:
            result_dict['Fuel capacity'] = fuel_capacity_elem.find_next_sibling('td').text.strip()

        # find Top Speed
        top_speed_elem = engine_data.find('td', text='Top Speed:')
        if top_speed_elem:
            result_dict['Top Speed'] = top_speed_elem.find_next_sibling('td').text.strip()

        # find Top Speed
        top_speed_elem = engine_data.find('td', text='Top Speed:')
        if top_speed_elem:
            result_dict['Top Speed'] = top_speed_elem.find_next_sibling('td').text.strip()

        # find Top Speed
        top_speed_elem = engine_data.find('td', text='Top Speed:')
        if top_speed_elem:
            result_dict['Top Speed'] = top_speed_elem.find_next_sibling('td').text.strip()        

        # find Acceleration 0-62 Mph (0-100 kph)
        acceleration_elem = engine_data.find('td', text='Acceleration 0-62 Mph (0-100 kph):')
        if acceleration_elem:
            result_dict['Acceleration 0-62 Mph (0-100 kph)'] = acceleration_elem.find_next_sibling('td').text.strip() 

        # find Drive Type
        drive_type_elem = engine_data.find('td', text='Drive Type:')
        if drive_type_elem:
            result_dict['Drive Type'] = drive_type_elem.find_next_sibling('td').text.strip() 

        # find Gearbox
        gearbox_elem = engine_data.find('td', text='Gearbox:')
        if gearbox_elem:
            result_dict['Gearbox'] = gearbox_elem.find_next_sibling('td').text.strip() 

        # find Front brakes
        front_elem = engine_data.find('td', text='Front:')
        if front_elem:
            result_dict['Front'] = front_elem.find_next_sibling('td').text.strip() 

        # find Rear brakes
        rear_elem = engine_data.find('td', text='Rear:')
        if rear_elem:
            result_dict['Rear'] = rear_elem.find_next_sibling('td').text.strip() 

        # find Tire Size
        tire_size_elem = engine_data.find('td', text='Tire Size:')
        if tire_size_elem:
            result_dict['Tire Size'] = tire_size_elem.find_next_sibling('td').text.strip()

        # find Length
        length_elem = engine_data.find('td', text='Length:')
        if length_elem:
            result_dict['Length'] = length_elem.find_next_sibling('td').text.strip() 

        # find Width
        width_elem = engine_data.find('td', text='Width:')
        if width_elem:
            result_dict['Width'] = width_elem.find_next_sibling('td').text.strip()

        # find Height
        height_elem = engine_data.find('td', text='Height:')
        if height_elem:
            result_dict['Height'] = height_elem.find_next_sibling('td').text.strip() 

        # find Front/rear Track
        front_rear_track_elem = engine_data.find('td', text='Front/rear Track:')
        if front_rear_track_elem:
            result_dict['Front/rear Track'] = front_rear_track_elem.find_next_sibling('td').text.strip() 

        # find Wheelbase
        wheelbase_elem = engine_data.find('td', text='Wheelbase:')
        if wheelbase_elem:
            result_dict['Wheelbase'] = wheelbase_elem.find_next_sibling('td').text.strip()

        # find Ground Clearance
        ground_clearance_elem = engine_data.find('td', text='Ground Clearance:')
        if ground_clearance_elem:
            result_dict['Ground Clearance'] = ground_clearance_elem.find_next_sibling('td').text.strip()

        # find Cargo Volume
        cargo_volume_elem = engine_data.find('td', text='Cargo Volume:')
        if cargo_volume_elem:
            result_dict['Cargo Volume'] = cargo_volume_elem.find_next_sibling('td').text.strip()

        # find Aerodynamics (Cd)
        aerodynamics_elem = engine_data.find('td', text='Aerodynamics (Cd):')
        if aerodynamics_elem:
            result_dict['Aerodynamics (Cd)'] = aerodynamics_elem.find_next_sibling('td').text.strip() 

        # find Turning circle (curb to curb)
        turning_circle_elem = engine_data.find('td', text='Turning circle (curb to curb):')
        if turning_circle_elem:
            result_dict['Turning circle (curb to curb)'] = turning_circle_elem.find_next_sibling('td').text.strip()

        # find Unladen Weight
        unladen_weight_elem = engine_data.find('td', text='Unladen Weight:')
        if unladen_weight_elem:
            result_dict['Unladen Weight'] = unladen_weight_elem.find_next_sibling('td').text.strip()

        # find Gross Weight Limit
        gross_weight_limit_elem = engine_data.find('td', text='Gross Weight Limit:')
        if gross_weight_limit_elem:
            result_dict['Gross Weight Limit'] = gross_weight_limit_elem.find_next_sibling('td').text.strip() 

        # find City
        city_elem = engine_data.find('td', text='City:')
        if city_elem:
            result_dict['City'] = city_elem.find_next_sibling('td').text.strip() 

        # find Highway
        highway_elem = engine_data.find('td', text='Highway:')
        if height_elem:
            result_dict['Highway'] = height_elem.find_next_sibling('td').text.strip() 

         # find Combined
        combined_elem = engine_data.find('td', text='Combined:')
        if combined_elem:
            result_dict['Combined'] = combined_elem.find_next_sibling('td').text.strip() 

        # find CO2 Emissions
        emissions_elem = engine_data.find('td', text='CO2 Emissions:')
        if emissions_elem:
            result_dict['CO2 Emissions'] = emissions_elem.find_next_sibling('td').text.strip()        

    # Find all images and retrieve URL
    all_images = soup.find_all('a', class_="s_gallery")
    images_urls = []
    for image in all_images:
        images_urls.append(image['href'])
    result_dict['image urls'] = images_urls
    
    return result_dict
```

```python
main_df_list = []

for url in model_url_list:
    df = get_model_data(url)
    main_df_list.append(df)
    
main_df = pd.DataFrame(main_df_list)
main_df.to_csv(OUTPUT_FILE, index=False)
print(len(main_df))
```
