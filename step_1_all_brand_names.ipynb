```python
from bs4 import BeautifulSoup
import requests
import pandas as pd

CSV_FILE_NAME = 'brand_names.csv'

# send a request to the website
url = 'https://www.autoevolution.com/cars/'
r = requests.get(url)
soup = BeautifulSoup(r.content, 'html.parser')

# create empty lists to store the data
brand_urls = []
brand_names = []
brand_logos = []

# find all the divs with class 'carman'
carman_divs = soup.find_all('div', class_='carman')

# loop through each div and extract the relevant information
for div in carman_divs:
    # extract brand url
    brand_url = div.find('a').get('href')
    brand_urls.append(brand_url)

    # extract brand name
    brand_name = div.find('span', itemprop='name').text.strip()
    brand_names.append(brand_name)

    # extract brand logo
    brand_logo = div.find('img', itemprop='logo').get('src')
    brand_logos.append(brand_logo)

# create a pandas dataframe with the extracted data
brands_df = pd.DataFrame({'Brand URL': brand_urls, 'Brand Name': brand_names, 'Brand Logo': brand_logos})

brands_df.to_csv(CSV_FILE_NAME, index=False) 
# print the len
len(f"Extracted total {len(brands_df)} brands. Saved as {CSV_FILE_NAME}")
```
